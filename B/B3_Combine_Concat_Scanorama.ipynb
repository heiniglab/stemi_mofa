## Data from B2 

```python
anndata_dict = dict.fromkeys(variants)
```

```python
for key in anndata_dict:
    dataset_path = data_path + variants_dictionary[key]
    
    print(dataset_path)
    print('Last modified' + time.ctime(os.path.getmtime(dataset_path)))
    
    adata_orig = sc.read_h5ad(dataset_path)
    
    anndata_dict[key]= adata_orig # save anndata in dictionary
    anndata_dict[key].uns['data_load_time'] = time.ctime(os.path.getmtime(dataset_path))  # save the last modified timestamp of the data loaded
    anndata_dict[key].uns['data_load_name'] = dataset_path # save the data path of the data loaded 
    
```

## Add new metadata to anndata

```python
for key in anndata_dict:
    anndata_dict_index = anndata_dict[key].obs.index   # Save index for adding afterwards
    
    #### Merge library_sample_group_mapping
    anndata_dict[key].obs = pd.merge(anndata_dict[key].obs, library_sample_group_mapping ,how="left", left_on=['library', 'display_name'], right_on=['library', 'sample']) 
    
    #### Merge library_sample_ef_mapping
    anndata_dict[key].obs = pd.merge(anndata_dict[key].obs, library_sample_ef_mapping ,how="left", left_on=['library', 'display_name'], right_on=['library', 'sample'])
    
    #### Restore index
    anndata_dict[key].obs.index = anndata_dict_index
```
